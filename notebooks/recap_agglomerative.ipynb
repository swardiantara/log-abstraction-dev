```python
import json
import pandas as pd

def build_recap(model, source_dir, output_dir):
    recap_overall = pd.DataFrame()
    embeddings = os.listdir(source_dir)
    for embedding in embeddings:
        embedding_path = os.path.join(source_dir, embedding)
        linkages = os.listdir(embedding_path)
        for linkage in linkages:
            linkage_path = os.path.join(embedding_path, linkage)
            thresholds = os.listdir(linkage_path)
            for threshold in thresholds:
                threshold_path = os.path.join(linkage_path, threshold)
                files = os.listdir(threshold_path)
                if len(files) < 1:
                    continue
                scenario_path = os.path.join(threshold_path, files[1])
                with open(scenario_path) as eval_report_file:
                    eval_report = json.load(eval_report_file)
                    silhouette = round(float(eval_report['silhouette_avg']), 5)
                    calinski_harabasz_avg = round(float(eval_report['calinski_harabasz_avg']), 5)
                    duration = eval_report['duration'].split()[0]
                    columns = ['model', 'embedding', 'linkage', 'threshold', 'silhouette', 'calinski', 'duration']
                    row_list = [model, embedding, linkage, threshold, silhouette, calinski_harabasz_avg, duration]
                    scenario_df = pd.DataFrame([row_list], index=None, columns=columns)
                # columns = ['dataset', 'encoder', 'n_layer', 'n_head', 'bidirectional', 'pooling', 'class_weight', 'scenario', 'best_epoch', 'accuracy', 'precision', 'recall', 'f1_score', 'confidence_avg', 'confidence_std']
                # row_list = [dataset, encoder, n_layer, n_head, bidirectional, pooling, weight, loss, best_epoch, accuracy, precision, recall, f1_score, confidence_avg, confidence_std]
                # scenario_df = pd.DataFrame([row_list], index=None, columns=columns)
                    recap_overall = pd.concat([recap_overall, scenario_df], ignore_index=True)
                    if not os.path.exists(output_dir):
                        os.makedirs(output_dir)
                    recap_overall.to_excel(os.path.join(output_dir, f'recap_experiment_{model}.xlsx'), index=False, float_format='%.5f')
    return recap_overall
```

```python
model = 'agglomerative'
source_dir = os.path.join('experiments', model)
output_dir = os.path.join('experiments', 'recap', model)
build_recap(model, source_dir, output_dir)
```
